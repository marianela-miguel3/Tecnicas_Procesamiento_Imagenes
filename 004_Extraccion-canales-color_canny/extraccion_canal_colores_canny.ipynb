```python
import numpy as np
#numpy es para matrices
import cv2
from cv2 import imread
#cv2 es para vision por computador y procesamiento de imagenes
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
plt.show()
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
#marco los bordes con canny
img_canny_color= cv2.Canny(img_rgb, 100, 150)
plt.imshow(img_canny_color)
```

```python
img_canny_color.shape
```

```python
# Convertir a escala de grises
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray, cmap='gray')
```

```python
img_canny_gray= cv2.Canny(img_gray, 100, 150)
plt.imshow(img_canny_gray, cmap='gray')
```

```python
# Mostrar ambas imágenes una al lado de la otra
plt.figure(figsize=(10, 5))

# Imagen a color
plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Color")
plt.axis('off')

# Imagen en escala de grises
plt.subplot(1, 2, 2)
plt.imshow(img_gray, cmap='gray')
plt.title("Escala de grises")
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Mostrar ambas imágenes una al lado de la otra
plt.figure(figsize=(10, 5))

# Imagen a color con canny
plt.subplot(1, 2, 1)
plt.imshow(img_canny_color)
plt.title("Color")
plt.axis('off')

# Imagen en escala de grises con canny
plt.subplot(1, 2, 2)
plt.imshow(img_canny_gray, cmap='gray')
plt.title("Escala de grises")
plt.axis('off')

plt.tight_layout()
plt.show()
```
