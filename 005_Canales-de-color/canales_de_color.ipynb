```python
import numpy as np
#numpy es para matrices
import cv2
from cv2 import imread
#cv2 es para vision por computador y procesamiento de imagenes
import matplotlib.pyplot as plt
```

```python
!wget https://www2.eecs.berkeley.edu/Research/Projects/CS/vision/bsds/BSDS300/html/images/plain/normal/color/317080.jpg
```

```python
img= imread('317080.jpg')
plt.imshow(img)
plt.show()
```

```python
img_canny_color= cv2.Canny(img, 100, 150)
plt.imshow(img_canny_color)
```

```python
img_only_0=img[:,:,0]
plt.imshow(img_only_0)
plt.show()
```

```python
img_only_1=img[:,:,1]
plt.imshow(img_only_1)
plt.show()
```

```python
img_only_2=img[:,:,2]
plt.imshow(img_only_2)
plt.show()
```

```python
ver = np.hstack((img_only_0, img_only_1, img_only_2))
```
