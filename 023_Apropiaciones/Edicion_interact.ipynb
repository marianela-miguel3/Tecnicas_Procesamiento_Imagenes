```python
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider, Button, Layout, VBox
from IPython.display import display
from datetime import datetime

!wget https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcQl-j6T2DPoM3_a4kfQrzvB2dtXrhvbaA-jVSKBWKmremwrJ9lPdGSRR9CmI7HVPXqnB_A&usqp=CAU
# Subir imagen
img_bgr = cv2.imread('imagen')
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# Variable global para guardar la imagen modificada
ultima_modificada = img_rgb.copy()

# Función de ajuste HSV
def ajustar_hsv(h_shift=0, s_mult=100, v_mult=100):
    global ultima_modificada
    s_mult = s_mult / 100
    v_mult = v_mult / 100

    img_hsv = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HSV).astype(np.float32)

    img_hsv[..., 0] = (img_hsv[..., 0] + h_shift) % 180
    img_hsv[..., 1] = np.clip(img_hsv[..., 1] * s_mult, 0, 255)
    img_hsv[..., 2] = np.clip(img_hsv[..., 2] * v_mult, 0, 255)

    img_hsv = img_hsv.astype(np.uint8)
    img_mod = cv2.cvtColor(img_hsv, cv2.COLOR_HSV2RGB)

    ultima_modificada = img_mod.copy()

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(img_rgb)
    plt.title("Imagen original")
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.imshow(img_mod)
    plt.title("Imagen modificada")
    plt.axis("off")
    plt.show()

# Botón para guardar la imagen
def guardar_imagen(b):
    nombre = f"imagen_modificada_{datetime.now().strftime('%H%M%S')}.png"
    img_bgr_save = cv2.cvtColor(ultima_modificada, cv2.COLOR_RGB2BGR)
    cv2.imwrite(nombre, img_bgr_save)
    files.download(nombre)

# Crear botón
boton_guardar = Button(description="Guardar imagen modificada", button_style='success')
boton_guardar.on_click(guardar_imagen)

# Mostrar controles
display(boton_guardar)

# Interactivos HSV
interact(
    ajustar_hsv,
    h_shift=IntSlider(min=-90, max=90, step=5, value=0, description="Hue", readout=True, layout=Layout(width='400px')),
    s_mult=IntSlider(min=0, max=300, step=10, value=100, description="Sat (%)", readout=True, layout=Layout(width='400px')),
    v_mult=IntSlider(min=0, max=300, step=10, value=100, description="Val (%)", readout=True, layout=Layout(width='400px')),
)
```
